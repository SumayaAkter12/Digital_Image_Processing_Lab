Read a grayscale image and perform the following operations

a. Compare different edge detection algorithms implemented in spatial domain.  

b. Illustrate the segmentation based on gray level.

c. Demonstrate the segmentation by adaptive thresholding.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the grayscale image
image_path = "cat.jpeg"  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found!")
    exit()

# Display the original image
plt.figure(figsize=(10, 10))
plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')







```

# a. Compare Edge Detection Algorithms

```python
# Sobel Edge Detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

plt.subplot(3, 3, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis('off')
```

```python
# Prewitt Edge Detection (using custom kernels)
kernel_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kernel_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
prewitt_x = cv2.filter2D(image, -1, kernel_x)
prewitt_y = cv2.filter2D(image, -1, kernel_y)
prewitt_combined = cv2.magnitude(prewitt_x.astype(np.float32), prewitt_y.astype(np.float32))

plt.subplot(3, 3, 3)
plt.imshow(prewitt_combined, cmap='gray')
plt.title("Prewitt Edge Detection")
plt.axis('off')
```

```python

# Laplacian of Gaussian (LoG) Edge Detection
log_edges = cv2.Laplacian(image, cv2.CV_64F)

plt.subplot(3, 3, 4)
plt.imshow(log_edges, cmap='gray')
plt.title("Laplacian of Gaussian (LoG)")
plt.axis('off')
```

 b. Segmentation Based on Gray Level

```python
# Global Thresholding (e.g., Otsu's Method)
_, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.subplot(3, 3, 6)
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu's Thresholding")
plt.axis('off')
```

# c. Segmentation by Adaptive Thresholding

```python
# Mean Adaptive Thresholding
mean_adaptive_thresh = cv2.adaptiveThreshold(
    image,
    maxValue=255,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C,
    thresholdType=cv2.THRESH_BINARY,
    blockSize=11,
    C=2,
)

plt.subplot(3, 3, 7)
plt.imshow(mean_adaptive_thresh, cmap='gray')
plt.title("Mean Adaptive Thresholding")
plt.axis('off')

```

```python
# Gaussian Adaptive Thresholding
gaussian_adaptive_thresh = cv2.adaptiveThreshold(
    image,
    maxValue=255,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType=cv2.THRESH_BINARY,
    blockSize=11,
    C=2,
)

plt.subplot(3, 3, 8)
plt.imshow(gaussian_adaptive_thresh, cmap='gray')
plt.title("Gaussian Adaptive Thresholding")
plt.axis('off')
```

```python
# Show all results together
plt.tight_layout()
plt.show()
```
